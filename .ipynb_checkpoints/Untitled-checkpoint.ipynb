```python
from fastai.collab import *
from fastai.tabular import *
path = Path("/home/sloth/datasets/anime/anime-recommendations-database")
```

```python
anime = pd.read_csv(path/"anime_clean.csv")
rating = pd.read_csv(path/"rating_clean.csv")
```

```python
anime_rating = anime.merge(rating, on="anime_id")
anime_rating.head()
```

```python
anime_rating = anime_rating.drop(["episodes", "name"], axis=1)
anime_rating.head()
```

```python
dep_var = 'user_rating'
cont_names = ["avg_rating", "members"]
cat_names = ['anime_id', 'Movie', 'TV', 'OVA', 'Special', 'Music', 'ONA', 'Other', "user_id", *[f'g{i}' for i in range(43)]]
procs = [Categorify, Normalize]
```

```python
test_user = anime_rating.loc[anime_rating['user_id'] == 50]
test_user
```

```python
anime_rating = anime_rating[anime_rating.user_id != 50]
anime_rating.head()
```

```python
test = TabularList.from_df(test_user, path=path, cat_names=cat_names, cont_names=cont_names)
```

```python
data = (TabularList.from_df(anime_rating, path=path, cat_names=cat_names, cont_names=cont_names, procs=procs)
                .split_by_rand_pct(0.1)
                .label_from_df(cols=dep_var, label_cls=FloatList)#log=True)
                .add_test(test)
                .databunch(bs=4096))
```

```python
data.show_batch()
```

```python
#max_log_y = np.log(10.05)
y_range = torch.tensor([0.5, 10.5], device=defaults.device)
```

```python
learn = tabular_learner(data, layers=[1000,500], ps=[0.001,0.01], emb_drop=0.04, 
                        y_range=y_range, metrics=exp_rmspe)
```

```python
learn.lr_find()
learn.recorder.plot(suggestion=True)
```

```python
torch.cuda.empty_cache()
```

```python
learn.fit_one_cycle(5,3.5e-3)
```

```python
learn.get_preds(DatasetType.Test())
```
