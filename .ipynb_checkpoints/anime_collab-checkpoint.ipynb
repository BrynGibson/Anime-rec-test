```python
from fastai.collab import *
from fastai.tabular import *
path = Path("/home/sloth/datasets/anime/anime-recommendations-database")
```

```python
from fastai_utils import cartesian_product, score
```

```python
anime = pd.read_csv(path/"anime_clean.csv")
rating = pd.read_csv(path/"rating_clean.csv")
```

```python
anime.head()
```

```python
rating.head()
```

```python
anime_rating = anime.merge(rating, on="anime_id")
anime_rating.head()
```

```python
# creating an exmaple user to make a prediction on
test_user = rating.loc[rating['user_id'] == 50]
test_user
```

```python
# removing test user from training set
rating = rating[rating.user_id != 50]
rating.head()
```

```python
data = CollabDataBunch.from_df(rating, seed=42, valid_pct=0.1, user_name="user_id", item_name="anime_id", rating_name="user_rating",bs=8192)
```

```python
data.show_batch()
```

```python
y_range = [0.5,10.5]
```

```python
torch.cuda.empty_cache()
```

```python
learn = collab_learner(data, n_factors=40, y_range=y_range, wd=1e-1, metrics=[root_mean_squared_error, mean_squared_error])
```

```python
learn.lr_find()
learn.recorder.plot(skip_end=15, suggestion=True)
```

```python
learn.fit_one_cycle(5, 3e-2)
```
